```python
def Get_Age_Group(age):

 if age <= 20:
  return "< 20"
 if age >20 and age <= 30:
    return "21-30"
 if age >30 and age <= 40:
    return "31-40"
 if age >40 and age <= 50: 
    return "41-50"
 if age > 50 and age <= 60:
    return "51-60"
 if age > 60 and age <= 70:
    return "61-70"
 if age >70 and age <= 80: 
    return "71-80"
 else:
    return "> 80"
```

```python
h_df['Age_group']=h_df['age'].apply(lambda x:Get_Age_Group(x))
```

```python
import matplotlib.pyplot as plt

# Group data by stroke and gender, calculating the mean for BMI, glucose level, and age
grouped_data = h_df.groupby(['stroke', 'gender'])[['bmi', 'avg_glucose_level', 'age']].mean().reset_index()

# Set up a color map for stroke status
stroke_colors = {0: 'skyblue', 1: 'lightcoral'}

# Set up a grid of subplots for each metric
fig, axes = plt.subplots(1, 3, figsize=(18, 6), sharey=True)

# Plot each metric in a separate subplot
for i, metric in enumerate(['bmi', 'avg_glucose_level', 'age']):
    # Create a pivoted DataFrame for plotting
    pivoted_data = grouped_data.pivot(index='gender', columns='stroke')[metric]
    
    # Plot bars for each stroke status with assigned colors
    bars = pivoted_data.plot(
        kind='bar', 
        color=[stroke_colors[status] for status in pivoted_data.columns],
        edgecolor='black', 
        ax=axes[i]
    )
    
    # Add labels on top of each bar
    for container in bars.containers:
        bars.bar_label(container, fmt='%.2f', label_type='edge', padding=3)
    
    
    axes[i].set_title(f'Average {metric.capitalize()} by Stroke and Gender')
    axes[i].set_xlabel('Gender')
    axes[i].set_ylabel('Average Value' if i == 0 else '')
    axes[i].legend(title='Stroke Status', labels=['No Stroke', 'Stroke'])
    axes[i].tick_params(axis='x', rotation=45)


plt.suptitle('Average BMI, Glucose Level, and Age by Stroke Status and Gender', fontsize=16)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
