```python
def Get_Age_Group(age):

 if age <= 20:
  return "< 20"
 if age >20 and age <= 30:
    return "21-30"
 if age >30 and age <= 40:
    return "31-40"
 if age >40 and age <= 50: 
    return "41-50"
 if age > 50 and age <= 60:
    return "51-60"
 if age > 60 and age <= 70:
    return "61-70"
 if age >70 and age <= 80: 
    return "71-80"
 else:
    return "> 80"
```

```python
h_df['Age_group']=h_df['age'].apply(lambda x:Get_Age_Group(x))# Age_group is new col
```
